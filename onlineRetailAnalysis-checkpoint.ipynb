```python
import pandas as pd
```

```python
df=pd.read_excel("Online Retail.xlsx")
```

```python
df.isnull().sum()
```

```python
# For removing missing values
df=df.dropna(subset=['CustomerID','Description'])
```

```python
df.shape
```

```python
#negative quantity values 
(df['Quantity'] < 0).sum()
```

```python
#cancelled transaction are
df['InvoiceNo'].str.startswith('C').sum()
```

```python
#removing missing CustomerID,Description,Cancelled/negative Quantity
df=df[df['Quantity'] > 0]
```

```python
df.shape
```

```python
#adding Revenue coloumn
df['Revenue']=df['Quantity']*df['UnitPrice']
```

```python
df.head()
```

```python
df['Month']=df['InvoiceDate'].dt.month
df['Year']=df['InvoiceDate'].dt.year
```

```python
df['MonthName']=df['InvoiceDate'].dt.month_name()
```

```python
df.head()
```

```python
#Total Revenue
df['Revenue'].sum()
```

```python
#Total Unique Customers
df['CustomerID'].nunique()
```

```python
#Total Orders
df['InvoiceNo'].nunique()
```

```python
#Average Revenue Per Order
df['Revenue'].sum()/df['InvoiceNo'].nunique()
```

```python
#Monthly Revenue Trend
monthly_revenue=df.groupby('MonthName')['Revenue'].sum()
monthly_revenue
```

```python
monthly_revenue=df.groupby('Month')['Revenue'].sum().sort_index()
monthly_revenue
```

```python
monthly_revenue=df.groupby(['Year','Month'])['Revenue'].sum().reset_index()
monthly_revenue
```

```python
country_revenue=df.groupby('Country')['Revenue'].sum().sort_values(ascending=False)
country_revenue.head(10)
```

```python
top_customers=df.groupby('CustomerID')['Revenue'].sum().sort_values(ascending=False).head(10)
top_customers
```

```python
top_products=df.groupby('Description')['Revenue'].sum().sort_values(ascending=False).head(10)
top_products
```

```python
df['Country'].value_counts().head(10)
```

```python
df.to_csv("cleaned_retail_final.csv",index=False)
```
